```python
##Using Correlation - to reduce highly correlated feature
import seaborn as sns
import matplotlib.pyplot as plt
#get correlations of each features in dataset
corrmat = df.corr()
top_corr_features = corrmat.index
plt.figure(figsize=(12,10))
#plot heat map
g=sns.heatmap(corrmat,annot=True,cmap="RdYlGn")
```

```python
#Get the Independent and Dependent Features
X=df.iloc[:,3:13]
Y=df.iloc[:,13]
```

```python
Y.value_counts()   ##The result shows us that it is a unbalanced dataset
```

```python
geography=pd.get_dummies(X['Geography'],drop_first=True)
geography.head()
```

```python
gender=pd.get_dummies(X['Gender'],drop_first=True)
gender.head()
```

```python
## Drop Categorical Features
X=X.drop(['Geography','Gender'],axis=1)
X.head()
```

```python
X=pd.concat([X,geography,gender],axis=1)
X.head()
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

XGBOOST Using Hyper parameter Optimization

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import xgboost
```

```python
##Stratified K-Fold cross validation is used to account for the unbalaced dataset
from sklearn.model_selection import StratifiedKFold
skfold = StratifiedKFold(n_splits=5)
```

```python
classifier=xgboost.XGBClassifier()
```

```python
random_search = RandomizedSearchCV(classifier,param_distributions=params,n_iter=5,scoring='roc_auc',n_jobs=-1,cv=skfold,verbose=3)
```

```python
random_search.fit(X_train,y_train)
```

#Using Artificial Neural Network(ANN)

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 50, kernel_initializer = 'truncated_normal',activation='relu',input_dim = 11))

# Adding the second hidden layer
classifier.add(Dense(units = 60, kernel_initializer = 'truncated_normal',activation='relu'))

# Adding the third hidden layer
classifier.add(Dense(units = 70, kernel_initializer = 'truncated_normal',activation='relu'))

# Adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'truncated_normal', activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
model_history=classifier.fit(X_train, y_train,validation_split=0.33, batch_size = 200, epochs = 50)
```

```python
import matplotlib.pyplot as plt 
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
score
```
